## Preprocessing

We now proceed to clean the message column in our data frame. 
Without the cleaning process, the dataset is often a cluster of words that the computer doesn't understand.

1)  First, we convert all text to lowercase, making our data more consistent and ensuring our model doesn't differentiate between identical words such as "Computer" and "computer". 

```python
df['message'] = df['message'].str.lower()
```

2) We remove the urls in the tweets and replace these simply with "url" as they do not give any information when we try to analyze text from words. 

```python
pattern_url = r'http[s]?://(?:[A-Za-z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9A-Fa-f][0-9A-Fa-f]))+'
subs_url = r'url-web'
df['message'] = df['message'].replace(to_replace = pattern_url, value = subs_url, regex = True)
```

3) Stopwords in NLTK are the most common words in any human language. By removing these words, we remove the low-level information from our text in order to give more focus to the important information.

```python
from nltk.corpus import stopwords

stopwords_list = stopwords.words('english')
print("Stopwords in the English language:\n\n", stopwords_list)
```

```python
without_stopwords = []
for tweet in range(len(df["message"])):
    split = df["message"][tweet].split(" ")
    for word in stopwords.words('english'):
        if word in split:
            split.remove(word)
    without_stopwords.append(' '.join(map(str, split)))
            
        
df["messages_without_noise"] = without_stopwords
```

4) The punctuation removal process will help to treat each text equally. For example, the word data and data! are treated equally after the process of removal of punctuations. Also punctuation such as $%&'()*+,-./:;<=>?@[\]^_`{|}~ will not be valuable when trying to predict our class

```python
import string
print(string.punctuation)
```

```python
# Function to remove any punctuation
def remove_punctuation(message):
    return ''.join([l for l in message if l not in string.punctuation])
```

```python
df["messages_without_noise"] = df["messages_without_noise"].apply(remove_punctuation)
```

5) Tokenization is breaking the raw text into small chunks. Tokenization breaks the raw text into a list of words called tokens. These tokens help in understanding the context or developing the model for the NLP

```python
from nltk.tokenize import word_tokenize, TreebankWordTokenizer

tokeniser = TreebankWordTokenizer()
df["tokens_messages_without_noise"] = df["messages_without_noise"].apply(tokeniser.tokenize)
```

6) Lemmatizing is the process of switching any word to its base word. We choose this method over its altenative, stemming, as lemmatization provides better results by producing real, dictionary words.

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
# Function to convert words to their lemmas
def tokenized_lemma(words, lemmatizer):
    return [lemmatizer.lemmatize(word) for word in words] 
```

```python
# Function to convert our tokens to a string
def tostring(x):
    return ' '.join([str(elem) for elem in x])
```

```python
df['lemma'] = df['tokens_messages_without_noise'].apply(tokenized_lemma, args=(lemmatizer, ))
df['lemma_string'] = df['lemma'].apply(tostring)
```

### Word Frequency

We first create a function that outputs the 25 least or most frequent words in a tweet:

```python
def frequency(tweet, type):
    # Count Vectorizer without stop words
    cv = CountVectorizer(stop_words = 'english')
    words = cv.fit_transform(tweet)
    
    # Count of words in the tweets and determine the frequency of each word
    sum_words = words.sum(axis = 0)
    words_frequency = [(word, sum_words[0, i]) for word, i in cv.vocabulary_.items()]
    words_frequency = sorted(words_frequency, key = lambda x: x[1], reverse = True)
    
    # Dataframe to stop top 25 words and their frequencies
    frequency = pd.DataFrame(words_frequency, columns = ['word','freq'])
    
    if type == "most":
        frequency = frequency.head(25)
    elif type == "least":
        frequency = frequency.tail(25)
    
    return frequency
```

We then use this function to define another function that will output a WordCloud diagram of the 25 least or most frequent words:

```python
def Word_cloud(df, type):
    
    #This applies the frequency function above to each sentiment and produces a string of the 25 most or least frequent words
    pro_words = tostring(list(frequency(df['lemma_string'][df["word_sentiments"] == 'Pro'], type)["word"]))
    anti_words = tostring(list(frequency(df['lemma_string'][df["word_sentiments"] == 'Anti'], type)["word"]))
    news_words = tostring(list(frequency(df['lemma_string'][df["word_sentiments"] == 'News'], type)["word"]))
    neutral_words = tostring(list(frequency(df['lemma_string'][df["word_sentiments"] == 'Neutral'], type)["word"]))
    
    # This string is then used to create a WordCloud
    pro_wordbank = WordCloud(width=800, 
                              height=500, 
                              random_state=73, 
                              max_font_size=110, 
                              background_color='white',
                              colormap="Greens").generate(pro_words)
    anti_wordbank = WordCloud(width=800, 
                               height=500, 
                               random_state=110, 
                               max_font_size=110, 
                               background_color='white',
                               colormap="Reds").generate(anti_words)
    news_wordbank = WordCloud(width=800, 
                              height=500, 
                              random_state=0, 
                              max_font_size=110, 
                              background_color='white',
                              colormap="Blues").generate(news_words)
    neutral_wordbank = WordCloud(width=800, 
                              height=500, 
                              random_state=10, 
                              max_font_size=110, 
                              background_color='white',
                              colormap="Oranges").generate(neutral_words)
    
    # Now we plot our WordCloud
    
    f, axarr = plt.subplots(2,2, figsize=(35,25))
    axarr[0,0].imshow(pro_wordbank, interpolation="bilinear")
    axarr[0,1].imshow(anti_wordbank, interpolation="bilinear")
    axarr[1,0].imshow(neutral_wordbank, interpolation="bilinear")
    axarr[1,1].imshow(news_wordbank, interpolation="bilinear")

    # Remove the ticks on the x and y axes
    for ax in f.axes:
        plt.sca(ax)
        plt.axis('off')

    axarr[0,0].set_title('Pro climate change\n', fontsize=35)
    axarr[0,1].set_title('Anti climate change\n', fontsize=35)
    axarr[1,0].set_title('Neutral\n', fontsize=35)
    axarr[1,1].set_title('News\n', fontsize=35)
    
    return
```

## 4. Data Engineering
<a class="anchor" id="1.1"></a>
<a href=#cont>Back to Table of Contents</a>

Before we move on we need to address an issue with our data- class imbalance.

Class imbalance occurs when the number of observations across different class labels are unevenly distributed. Why is this a problem? Most machine learning algorithms assume data equally distributed. So when we have a class imbalance, the machine learning classifier tends to be more biased towards the majority class, causing bad classification of the minority class.

So how do we solve this problem ? The answer is resampling. Put simply, resampling methods involve modifying the number of observations in each class. 

From our data we see that our pro class (1) out weighs the other classes by over 50%:

```python
df['sentiment'].value_counts()
```

To avoid this becoming an issue in our model we will apply a combination of upsampling (taking repeated random samples from the minority class until we have as many observations as the majority class) and downsampling (taking a random subset of the majority class small enough to match the number of observations in the minority class). We will set our class size to be approximately half of our majority class, 4250. 

```python
class_size = 4250
```

Next we separate our majority and minority classes:

```python
one = df[df['sentiment']== 1]
two = df[df['sentiment']== 2]
zero = df[df['sentiment']== 0]
minus_one = df[df['sentiment']== -1]
```

We start by downsampling our majority class:

```python
one_resampled = resample(one,
                          replace=False, # sample without replacement (no need to duplicate observations)
                          n_samples=class_size, # match number in minority class
                          random_state=42)
```

Now we upsample the minority classes: 

```python
two_resampled = resample(two,
                          replace=True, # sample with replacement (we need to duplicate observations)
                          n_samples=class_size, # match number in minority class
                          random_state=42) # reproducible results

zero_resampled = resample(zero,
                          replace=True, # sample with replacement (we need to duplicate observations)
                          n_samples=class_size, # match number in minority class
                          random_state=42) # reproducible results

minus_one_resampled = resample(minus_one,
                          replace=True, # sample with replacement (we need to duplicate observations)
                          n_samples=class_size, # match number in minority class
                          random_state=42) # reproducible results
```

Now we combine the upsampled and downsampled data: 

```python
resampled = pd.concat([one_resampled, two_resampled, zero_resampled, minus_one_resampled])
```

```python
resampled["sentiment"].value_counts()
```

Next we try only upsampling to our majority class:

```python
one_upsampled = one


two_upsampled = resample(two,
                          replace=True, # sample with replacement (we need to duplicate observations)
                          n_samples=8530, # match number in majority class
                          random_state=42) # reproducible results

zero_upsampled = resample(zero,
                          replace=True, # sample with replacement (we need to duplicate observations)
                          n_samples=8530, # match number in majority class
                          random_state=42) # reproducible results

minus_one_upsampled = resample(minus_one,
                          replace=True, # sample with replacement (we need to duplicate observations)
                          n_samples=8530, # match number in majority class
                          random_state=42) # reproducible results

upsampled = pd.concat([one_upsampled, two_upsampled, zero_upsampled, minus_one_upsampled])
```

Next we try only downsampling to our minority class:

```python
one_downsampled = resample(one,
                          replace=True, # sample without replacement (no need to duplicate observations)
                          n_samples=1296, # match number in minority class
                          random_state=42) # reproducible results


two_downsampled = resample(two,
                          replace=True, # sample without replacement (no need to duplicate observations)
                          n_samples=1296, # match number in minority class
                          random_state=42) # reproducible results

zero_downsampled = resample(zero,
                          replace=True, # sample without replacement (no need to duplicate observations)
                          n_samples=1296, # match number in minority class
                          random_state=42) # reproducible results

minus_one_downsampled = minus_one

downsampled = pd.concat([one_downsampled, two_downsampled, zero_downsampled, minus_one_downsampled])
```

## 6. Model Performance
<a class="anchor" id="1.1"></a>
<a href=#cont>Back to Table of Contents</a>

Below is a function that will test our model performance by producing the accuracy, precision, recall and F1 scores: 

```python
def scores(lm, X_test, y_test):
    # your code here
    accuracy = accuracy_score(y_test, lm.predict(X_test) )
    precision = precision_score(y_test, lm.predict(X_test), average = "weighted" )
    recall = recall_score(y_test, lm.predict(X_test), average = "weighted" )
    f = f1_score(y_test, lm.predict(X_test), average = "weighted" ) 
    return {"Accuracy": round(accuracy,3), "Precision":round(precision,3), "Recall": round(recall,3), "f1": round(f,3)}
```

```python
models = [lm, svc, knn, mnbc, tree, forest, abc]
model_names = ["Logistic Regression", "Linear Support Vector Classifier", "K Nearest Neighbours",
                   "Multinomial Naive Bayes Classifier", "Decision Tree", "Random Forest", "AdaBoostClassifier" ] 

for i in range(len(models)):
    print(model_names[i],": ", scores(models[i], X_test, y_test))
    print("")
```

To make this information a bit more paletable we can express the f1 scores graphically: 

```python
lm_f1 = list(scores(lm, X_test, y_test).items())[3][1]
svc_f1 = list(scores(svc, X_test, y_test).items())[3][1]
knn_f1 = list(scores(knn, X_test, y_test).items())[3][1]
mnbc_f1 = list(scores(mnbc, X_test, y_test).items())[3][1]
tree_f1 = list(scores(tree, X_test, y_test).items())[3][1]
forest_f1 = list(scores(forest, X_test, y_test).items())[3][1]
abc_f1 = list(scores(abc, X_test, y_test).items())[3][1]
```

```python
ax = plt.subplot(111)
ax.barh(model_names, [lm_f1, svc_f1, knn_f1, mnbc_f1, tree_f1, forest_f1, abc_f1])
ax.set_xlabel("f1 score")
for index, value in enumerate([round(value,2) for value in [lm_f1, svc_f1, knn_f1, mnbc_f1, tree_f1, forest_f1, abc_f1]]):
    plt.text(value, index, str(value))
```
